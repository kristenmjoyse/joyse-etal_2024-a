```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

CCdata = pd.read_excel("CheesequakeCreekTG.xlsm")
RRdata = pd.read_excel("RaritanRiverTG.xlsm")

CCdata.head()
RRdata.head()
```

```python
fig = plt.figure(figsize = (25,10))

# cc = fig.add_subplot(211)
rr = fig.add_subplot(111)

ccx = CCdata['Date']
ccy = CCdata['Water Level (m)']

# cc.plot(ccx, ccy, color = 'cornflowerblue', lw = 0.8)

rrx = RRdata['Date']
rry = RRdata['Water Level (m)']

rr.plot(rrx, rry, color = 'cornflowerblue', lw = 1.5, label = 'Raritan River')
rr.plot(ccx, ccy, color = 'mediumaquamarine', lw = 1.5, label = 'Cheesequake Creek')

rr.set_xlim(1976, 1977)
rr.tick_params(labelsize = 14)
rr.legend(fontsize = 16, frameon = False)

plt.savefig("RiverGauges.pdf", bbox_inches = 'tight', format = 'pdf')
```
