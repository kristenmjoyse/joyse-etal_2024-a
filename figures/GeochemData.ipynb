```python
from pathlib import Path
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_excel("geochem.xlsx", sheet_name = 'Sheet1')

data.head()
```

```python
fig = plt.figure(figsize = (6,10))

d13C = fig.add_subplot(131)
CN = fig.add_subplot(132)
d15N = fig.add_subplot(133)

data.plot(kind = 'line', x = 'd13C', y = 'Plotting Depth (cm)', color = 'black', linewidth = 1, marker = 'o', markerfacecolor = 'white', ax = d13C, legend = None)

d13C.invert_yaxis()
d13C.set_ylim(120,0)
d13C.set_xlim(-35, -15)
d13C.set_ylabel('Depth (cm)', fontsize = 14)
d13C.set_xlabel('Organic \u03B413C (\u2030)', fontsize = 14)
d13C.tick_params(axis = 'x', labelsize = 12)
d13C.tick_params(axis = 'y', labelsize = 12)

data.plot(kind = 'line', x = 'C/N', y = 'Plotting Depth (cm)', color = 'black', linewidth = 1, marker = 'o', markerfacecolor = 'white', ax = CN, legend = None)

CN.invert_yaxis()
CN.set_ylim(120,0)
CN.set_xlim(0,20)
CN.set_yticks([])
CN.set_xlabel('C/N', fontsize = 14)
CN.tick_params(axis = 'x', labelsize = 12)

data.plot(kind = 'line', x = 'd15N', y = 'Plotting Depth (cm)', color = 'black', linewidth = 1, marker = 'o', markerfacecolor = 'white', ax = d15N, legend = None)

d15N.invert_yaxis()
d15N.set_ylim(120,0)
d15N.set_xlim(-3,9)
d15N.set_yticks([])
d15N.set_xlabel('\u03B415N (\u2030)', fontsize = 14)
d15N.tick_params(axis = 'x', labelsize = 12)

plt.savefig("GeochemData.pdf", bbox_inches = 'tight')
```
