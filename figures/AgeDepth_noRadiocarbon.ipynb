```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_excel("Age_Depth.xlsx", sheet_name = "All")

data.head(14)
```

```python
mdepth = data['Mean Depth (cm)']
mage = data['Mean Age (CE)']

deptherr = data['Depth Error (cm)']
ageerr = data ['Age Error (CE)']
```
